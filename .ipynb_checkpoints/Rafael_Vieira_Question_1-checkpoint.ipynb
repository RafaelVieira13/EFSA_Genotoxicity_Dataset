__c. Provide the same information in a HTML table format for substances for which EFSA has published an opinion after 2013.__ 

Let's create a new dataset that includes only the substances for which EFSA has published an opinion after 2013.

```python
def table_ex_1b(df):
    # Get the CAS_number and drop duplicated values
    CAS_number = df[['SUB_NAME','SUB_CASNUMBER']]
    CAS_number = CAS_number.drop_duplicates(subset='SUB_NAME')

    # Get the total number of assays performed
    assays = df.groupby('SUB_NAME').agg(Assays_performed=('SUB_NAME','count')).reset_index()
    
    # Get the results
    results = df.groupby(['SUB_NAME','RESULTS']).size().unstack(fill_value=0)
    results.reset_index(inplace=True)
    results = results[['SUB_NAME','Positive','Negative']]
    results

    # Merge Every Dataset
    merged_df_1 = pd.merge(assays, param[['SUB_NAME', 'termCode']], on='SUB_NAME', how='left')
    merged_df_2 = pd.merge(merged_df_1, CAS_number[['SUB_NAME', 'SUB_CASNUMBER']], on='SUB_NAME', how='left')
    df_1b = pd.merge(merged_df_2, results[['SUB_NAME','Positive','Negative']], on='SUB_NAME', how='left')

    #Rename Columns
    df_1b.rename(columns={
        'SUB_NAME':'Substance Name',
        'SUB_CASNUMBER':'Substance CAS Number',
        'Assays_performed':'Total Number of Assays Performed',
        'Positive':'Number of Positive Results',
        'Negative':'Number of Negative Results'
    }, inplace=True)

    #Change the columns position
    df_1b = df_1b[['Substance Name','termCode','Substance CAS Number','Total Number of Assays Performed','Number of Positive Results','Number of Negative Results']]

    # Display the dataset
    return df_1b
```
